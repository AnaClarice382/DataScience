```python
def get_duplicate_values(df, columnNameApp):
    duplicate = []
    unique = []
    for row in df[columnNameApp]:
        if row in unique:
            duplicate.append(row)
        else:
            unique.append(row)
    return duplicate
```

```python
def get_max_reviews(df, columnNameApp, columnReviewsApp):
    reviews_max = {}

    for app, review in zip(df[columnNameApp], df[columnReviewsApp]):
        if app in reviews_max and reviews_max[app] < review:
            reviews_max[app] = review
        if app not in reviews_max:
            reviews_max[app] = review
    return reviews_max
```

```python
reviews_max_google = get_max_reviews(google_store, 'App', 'Reviews')
```

```python
reviews_max_google['Instagram'] #check if its correct 
```

```python
def get_clean_data(df, indexAppName, indexReview, review_max_list):
    clean_data = []
    already_added = []

    for row in df.to_numpy():
        app = row[indexAppName]
        review = row[indexReview]
        if review == review_max_list[app] and app not in already_added:
            clean_data.append(row)
            already_added.append(app)
    return clean_data
        
```

```python
android_clean = get_clean_data(google_store, 0, 3, reviews_max_google)
```

```python
def is_english(app_name):
    non_ascii = 0
    for character in app_name:
        if ord(character) > 127:
            non_ascii += 1
    if non_ascii > 3:
        return False
    else:
        return True
```

```python
def get_only_english(df,indexAppName):
    clean_data = []
    for row in df:
        if(is_english(row[indexAppName])):
            clean_data.append(row)
    return clean_data
```

```python
def getFreeApps(df, indexColumnPrice):
    free_app = []
    for row in df:
        price = row[indexColumnPrice]
        if(price == '0'): #unfortunately there's the character $ in price
            free_app.append(row)
        elif(price == 0.0):
            free_app.append(row)
    return free_app
```
